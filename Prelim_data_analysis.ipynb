```python
import math
import operator

def entropy(given_dict):
    H=0
    #sorted_dict = sorted(given_dict.items(), key=operator.itemgetter(1),reverse=True)
    count=0
    per_elem=0
            
    for elem in given_dict:
        count+=given_dict[elem]
        per_elem+= math.log(given_dict[elem])*given_dict[elem]        
    return (math.log(count)- per_elem/count)/(math.log(2))

def cond_entropy(dict1, dict2):
    N_dict1=sum([dict1[elem] for elem in dict1])
    N_dict2=sum([dict2[elem] for elem in dict2])
    prob_dict1={}
    prob_dict2={}
    
    for elem in dict1:
        prob_dict1[elem]= dict1[elem]/N_dict1
    for elem in dict2:
        prob_dict2[elem]= dict2[elem]/N_dict2

    cond_H=0
    for elem in prob_dict2:
        cond_H+= - (prob_dict2[elem])*math.log(prob_dict2[elem]/prob_dict1[elem[:-1]])/math.log(2)
    
    return cond_H

def cond_entropy_2(dict1, dict2):
    N_dict1=sum([dict1[elem] for elem in dict1])
    N_dict2=sum([dict2[elem1][elem2] for elem1 in dict2 for elem2 in dict2[elem1]])
    prob_dict1={}
    prob_dict2={}
    
    for elem in dict1:
        prob_dict1[elem]= dict1[elem]/N_dict1
    for elem1 in dict2:
        if elem1 not in prob_dict2:
            prob_dict2[elem1]={}
        for elem2 in dict2[elem1]:
            prob_dict2[elem1][elem2]= dict2[elem1][elem2]/N_dict2
        

    cond_H=0
    for elem1 in prob_dict2:
        for elem2 in prob_dict2[elem1]:
            cond_H+= - (prob_dict2[elem1][elem2])*math.log(prob_dict2[elem1][elem2]/prob_dict1[elem1])/math.log(2)
                                                                                               
    return cond_H

def cond_entropy_3(dict1, dict2):
    N_dict1=sum([dict1[elem1][elem2] for elem1 in dict1 for elem2 in dict1[elem1]])
    N_dict2=sum([dict2[elem1][elem2][elem3] for elem1 in dict2 for elem2 in dict2[elem1] for elem3 in dict2[elem1][elem2]])
    prob_dict1={}
    prob_dict2={}
    
    for elem1 in dict1:
        if elem1 not in prob_dict1:
            prob_dict1[elem1]={}
        for elem2 in dict1[elem1]:
            prob_dict1[elem1][elem2]= dict1[elem1][elem2]/N_dict1
            
    for elem1 in dict2:
        if elem1 not in prob_dict2:
            prob_dict2[elem1]={}
        for elem2 in dict2[elem1]:
            if elem2 not in prob_dict2[elem1]:
                prob_dict2[elem1][elem2]={}
            for elem3 in dict2[elem1][elem2]:
                prob_dict2[elem1][elem2][elem3]= dict2[elem1][elem2][elem3]/N_dict2
        

    cond_H=0
    for elem1 in prob_dict2:
        for elem2 in prob_dict2[elem1]:
            for elem3 in prob_dict2[elem1][elem2]:
                cond_H+= - (prob_dict2[elem1][elem2][elem3])*math.log(prob_dict2[elem1][elem2][elem3]/prob_dict1[elem1][elem2])/math.log(2)
                                                                                               
    return cond_H

test_dict={'a':10, 'b':10,'c':10, 'd':10}

stats_dict={}
features=['uni-entropy','bi-entropy','bi-rev-entropy','tri-entropy','tri-rev-entropy','cond-bi-entropy','cond-tri-entropy',\
          'neighbourhood-entropy','cond-bi-rev-entropy','cond-tri-rev-entropy','uni-w-entropy','bi-w-entropy','tri-w-entropy',\
         'cond-bi-w-entropy','cond-tri-w-entropy','neighbourhood-w-entropy']


for file in char_dist:
    stats_dict[file]={}
    stats_dict[file]['uni-entropy']=entropy(char_dist[file]['uni'])
    stats_dict[file]['bi-entropy']=entropy(char_dist[file]['bi'])
    stats_dict[file]['tri-entropy']=entropy(char_dist[file]['tri'])
    stats_dict[file]['uni-w-entropy']= entropy(word_dist_2[file]['uni'])
    stats_dict[file]['bi-w-entropy']= entropy(word_dist_2[file]['bi'])
    stats_dict[file]['tri-w-entropy']= entropy(word_dist_2[file]['tri'])
    
    stats_dict[file]['cond-bi-entropy']= cond_entropy(char_dist[file]['uni'], char_dist[file]['bi'])
    stats_dict[file]['cond-tri-entropy']= cond_entropy(char_dist[file]['bi'], char_dist[file]['tri'])
    stats_dict[file]['cond-bi-w-entropy']= cond_entropy_2(word_dist[file]['uni'], word_dist[file]['bi'])
    stats_dict[file]['cond-tri-w-entropy']= cond_entropy_3(word_dist[file]['bi'], word_dist[file]['tri'])
    
    
#     stats_dict[file]['bi-rev-entropy']=entropy(char_dist[file]['bi-rev'])
#     stats_dict[file]['tri-rev-entropy']=entropy(char_dist[file]['tri-rev'])
#     stats_dict[file]['neighbourhood-entropy']=entropy(char_dist[file]['neighbourhood'])
#     stats_dict[file]['cond-bi-entropy']= cond_entropy(char_dist[file]['uni'], char_dist[file]['bi'])
#     stats_dict[file]['cond-tri-entropy']= cond_entropy(char_dist[file]['bi'], char_dist[file]['tri'])
#     stats_dict[file]['cond-bi-rev-entropy']= cond_entropy(char_dist[file]['uni'], char_dist[file]['bi-rev'])
#     stats_dict[file]['cond-tri-rev-entropy']= cond_entropy(char_dist[file]['bi-rev'], char_dist[file]['tri-rev'])
```

```python
# Let c be the cipher text of length m, v is vocabulary of v tokens, and bigram model b

#text='he is a good guy he likes to play in the sun he has a wonderful body he is so good he likes to be a good guy'
text="vp gl k chhz cfd vp mgnpl sh jmkd gy svp lfy vp vkl k ohyzpaxfm thzd vp gl lh chhz vp mgnpl sh tp k chhz cfd"

import matplotlib.pyplot as plt
def normalize(counts):
    sub_dict={}
    for elem in counts:
        sub_dict[elem]={}
        cnt=sum([counts[elem][i] for i in counts[elem]])
        for i in counts[elem]:
            sub_dict[elem][i]=counts[elem][i]/cnt
    return sub_dict

def hope_fluctuate(prev_char_dist, iter_):
    same=0
    diff=0
    for elem in prev_char_dist[iter_]:
        if prev_char_dist[iter_][elem]==prev_char_dist[iter_-1][elem]:
            same+=1
        else:
            diff+=1
    return diff/(diff+same)


def EM_algo(text, prob_vc_dist,substitute_table):
    m= len(text)
    ciphers=list(set([i for i in text]))
    prev_char_dist={}
    counts={}
    Q={}
    R={}

    for v in prob_vc_dist['uni']:
        counts[v]={}
        Q[v]={}
        R[v]={}
        for cipher in ciphers:
            counts[v][cipher]=0
        for i in range(0,m):
            Q[v][i]=0
        for i in range(m-1,-1,-1):
            R[v][i]=0

    for v in prob_vc_dist['uni']:
        try:
            Q[v][0]=prob_vc_dist['bi'][' '+v]
        except Exception as e:
            Q[' '][0]=0
        try:
            R[v][m-1]=prob_vc_dist['bi'][v+' ']
        except Exception as e:
            R[' '][m-1]=0

    for j in range(1,m):        
        for i in prob_vc_dist['uni']:
            for k in prob_vc_dist['uni']:
                try:
                    Q[i][j]+=Q[k][j-1]*prob_vc_dist['bi'][k+i]*\
                    substitute_table[k][text[j-1]]
                except Exception as e:
                    continue

    for j in range(m-2,-1,-1):
        for i in prob_vc_dist['uni']:
            for k in prob_vc_dist['uni']:
                try:
                    R[i][j]+=R[k][j+1]*prob_vc_dist['bi'][i+k]*\
                    substitute_table[k][text[j+1]]
                except Exception as e:
                    continue

    for j in range(0,m):
        for i in prob_vc_dist['uni']:
            try:
                counts[i][text[j]]+=Q[i][j]*R[i][j]*substitute_table[i][text[j]]
            except:
                import pdb
                pdb.set_trace()

    substitute_table= normalize(counts)
    return substitute_table
    
    

def batch_EM_algo(text):
    m=len(text)
    ciphers=list(set([i for i in text]))
    clen=len(ciphers)-1
    vlen=len(vc_char_dist['uni'])-1
    substitute_table={}
    for v in vc_char_dist['uni']:
        substitute_table[v]={}
        for cipher in ciphers:
            if v==' ' or cipher==' ':
                substitute_table[v][cipher]=0
            else:
                substitute_table[v][cipher]=1/(clen)
    substitute_table[' '][' ']=1
    prob_vc_dist={}

    for key in vc_char_dist:
        prob_vc_dist[key]={}
        cnt= sum([vc_char_dist[key][elem] for elem in vc_char_dist[key]])
        for elem in vc_char_dist[key]:
            prob_vc_dist[key][elem]=vc_char_dist[key][elem]/cnt
    sents= text.split('\n')
    
    for iter_ in range(0,21):
        for sent in sents:
            substitute_table= EM_algo(sent,prob_vc_dist, substitute_table)
            
```

```python
## design the batch EM algorithm

# Let c be the cipher text of length m, v is vocabulary of v tokens, and bigram model b

import matplotlib.pyplot as plt

def normalize_with_smoothing(counts,clen):
    sub_dict={}
    val=0.000000000001
    for elem in counts:
        sub_dict[elem]={}
        cnt=sum([counts[elem][i] for i in counts[elem]])
        for i in counts[elem]:
            sub_dict[elem][i]=(counts[elem][i]+val)/(cnt+clen*val)
    return sub_dict


def normalize(counts):
    sub_dict={}
    for elem in counts:
        sub_dict[elem]={}
        cnt=sum([counts[elem][i] for i in counts[elem]])
        for i in counts[elem]:
            try:
                sub_dict[elem][i]=counts[elem][i]/cnt
            except Exception as e:
                print(elem, counts)
                
    return sub_dict

def hope_fluctuate(prev_char_dist, iter_):
    same=0
    diff=0
    for elem in prev_char_dist[iter_]:
        if prev_char_dist[iter_][elem]==prev_char_dist[iter_-1][elem]:
            same+=1
        else:
            diff+=1
    return diff/(diff+same)


def EM_algo(text, prob_vc_dist,substitute_table):
    m= len(text)
    ciphers=list(set([i for i in text]))
    prev_char_dist={}
    counts={}
    Q={}
    R={}

    for v in prob_vc_dist['uni']:
        counts[v]={}
        Q[v]={}
        R[v]={}
        for cipher in ciphers:
            counts[v][cipher]=0
        for i in range(0,m):
            Q[v][i]=0
        for i in range(m-1,-1,-1):
            R[v][i]=0

    for v in prob_vc_dist['uni']:
        try:
            Q[v][0]=prob_vc_dist['bi'][' '+v]
        except Exception as e:
            Q[' '][0]=0
        try:
            R[v][m-1]=prob_vc_dist['bi'][v+' ']
        except Exception as e:
            R[' '][m-1]=0

    for j in range(1,m):        
        for i in prob_vc_dist['uni']:
            for k in prob_vc_dist['uni']:
                try:
                    Q[i][j]+=Q[k][j-1]*prob_vc_dist['bi'][k+i]*\
                    substitute_table[k][text[j-1]]
                except Exception as e:
                    continue

    for j in range(m-2,-1,-1):
        for i in prob_vc_dist['uni']:
            for k in prob_vc_dist['uni']:
                try:
                    R[i][j]+=R[k][j+1]*prob_vc_dist['bi'][i+k]*\
                    substitute_table[k][text[j+1]]
                except Exception as e:
                    continue

    for j in range(0,m):
        for i in prob_vc_dist['uni']:
            try:
                counts[i][text[j]]+=Q[i][j]*R[i][j]*math.pow(substitute_table[i][text[j]],3)
            except:
                import pdb
                pdb.set_trace()

    #substitute_table= normalize(counts)
    return counts
    
    

def batch_EM_algo(text,en_flag):
    m=len(text)
    print(m)
    text= re.sub("[0-9!@#$%^&*\(\)\_\+-=:;\\'>.<.,?/|\n]+", "", text)
    if en_flag:
        text=re.sub("[^a-zA-Z ]+",'',text)
    ciphers=list(set([i for i in text]))
#     print(ciphers)
    clen=len(ciphers)-1
    vlen=len(vc_char_dist['uni'])-1
    substitute_table={}
    for v in vc_char_dist['uni']:
        substitute_table[v]={}
        for cipher in ciphers:
            if v==' ' or cipher==' ':
                substitute_table[v][cipher]=0
            else:
                substitute_table[v][cipher]=1/(clen)
    substitute_table[' '][' ']=1
    prob_vc_dist={}

    for key in vc_char_dist:
        prob_vc_dist[key]={}
        cnt= sum([vc_char_dist[key][elem] for elem in vc_char_dist[key]])
        for elem in vc_char_dist[key]:
            prob_vc_dist[key][elem]=vc_char_dist[key][elem]/cnt
            

    words= text.split(" ")
    n_words=len(words)
    
    for iter_ in range(0,1):
        print(iter_)
        prev_counts={}
        start=0
#         print(n_words)
        while True:
            if start+10>=(n_words):
                break
            sent=" ".join([w for w in words[start:start+20]])
            counts= EM_algo(sent,prob_vc_dist, substitute_table)
            for i in counts:
                if i not in prev_counts:
                    prev_counts[i]={}
                for j in counts[i]:
                    if j not in prev_counts[i]:
                        prev_counts[i][j]=0
                    prev_counts[i][j]+=counts[i][j]
            start+=10
                    
        substitute_table= normalize(prev_counts)
#         for elem in ciphers:
#             if elem not in substitute_table['C']:
#                 continue
#             if substitute_table['C'][elem]> substitute_table['V'][elem]:
#                 print(elem, 'C')
#             elif substitute_table['V'][elem]> substitute_table['C'][elem]:
#                 print(elem, 'V')
#             else:
#                 print(elem, 'undecided')
        
    return substitute_table,ciphers
```
